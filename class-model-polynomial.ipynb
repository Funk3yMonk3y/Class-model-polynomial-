```python
import math
import cmath
class Polynom:
    def __init__(self,a=1,b=1,c=1):
        self.a = a;
        if self.a == 0:
            self.a = 1
            print('a = 0 error')
        self.b = b;
        self.c = c;  
    def D(self):
        return self.b**2 -4*self.a*self.c
    
    def reell(self):
        if self.D()== 0:
            return 1
        elif self.D()>0:
            return 2
        elif self.D()<0:
            return 0
    
    def komplex(self):
        if self.D()==0:
            return 0
        if self.D()>0:
            return 0
        if self.D()<0:
            return 2
    
    def nullstellen(self):
        a = self.a
        b = self.b
        c = self.c
        if self.D()== 0:
            return (-b/(2*a))
        else:
            z = (cmath.sqrt(self.D())/(2 * a))
            return ((-b)/(2*a) + z,((-b)/(2*a) - z))
    
    def diff(self):
        a = self.a
        b = self.b
        c = self.c
        return (2*a,b)
    
    def __str__(self):
        return str(self.a) + "*x^2 + " + str(self.b) +"*x + "+ str(self.c)
    
    def __eq__(self,pol2):
        if self.a == pol2.a and self.b == pol2.b and self.c == pol2.c:
            return True
        else:
            return False
    
        
p = (Polynom(4,-12,-40))
q = (Polynom(1,12,37))
print(q)
print(p)
print(q.diff())
print(p.diff())
print(q==p)
print(q==q)
print(p.nullstellen())
print(q.nullstellen())
```
